```python
no_nulls = set(data.columns[np.sum(data.isnull()) == 0]) #Provide a set of columns with 0 missing values.
print(no_nulls)
```

```python
#  Which columns have the most missing values?
```

```python
most_25_missing_cols = set(data.columns[np.sum(data.isnull())/data.shape[0] > .25]) #Provide a set of columns with more than 25% of the values missing
most_50_missing_cols = set(data.columns[np.sum(data.isnull())/data.shape[0] > .50]) #Provide a set of columns with more than 50% of the values missing
most_75_missing_cols = set(data.columns[np.sum(data.isnull())/data.shape[0] > .75]) #Provide a set of columns with more than 75% of the values missing

print('The columns with more than 25% of the values: ', most_25_missing_cols)
print('The columns with more than 50% of the values: ', most_50_missing_cols)
print('The columns with more than 75% of the values: ', most_75_missing_cols)
```
